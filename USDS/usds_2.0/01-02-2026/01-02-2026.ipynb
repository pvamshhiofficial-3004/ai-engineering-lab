# File Operations

```python
with open("temp/temp_1.txt", "r") as f:
    print(f.tell()) # tells the current location of pointer
    all_contents = f.read()
    print(f.tell()) # tells the current location of pointer
    print(all_contents)
```

```python
with open("temp/temp_1_copy.txt", "w") as f:
    print(f.tell()) # tells the current location of pointer
    # all_contents = f.read() # read will not work in write mode
    f.write("Welcome to data science bootcamp")
    print(f.tell()) # tells the current location of pointer
```

```python
with open("temp/temp_1_copy.txt", "a") as f:
    print(f.tell()) # tells the current location of pointer
    # all_contents = f.read() # read will not work in write mode
    f.write(" -- Welcome to data science bootcamp")
    print(f.tell()) # tells the current location of pointer
```

```python
with open("temp/temp_1_copy.txt", "a") as f:
    print(f.tell()) # tells the current location of pointer
    # all_contents = f.read() # read will not work in write mode
    f.write(" \n-- Welcome to data science bootcamp")
    print(f.tell()) # tells the current location of pointer
```

```python
with open("temp/temp_1_copy.txt", "r+") as f:
    print(f.tell()) # tells the current location of pointer
    # all_contents = f.read() # read will not work in write mode
    # print(all_contents)
    f.write("*****")
    print(f.tell()) # tells the current location of pointer
```

```python
with open("temp/temp_1_copy_2.txt", "a+") as f:
    print(f.tell()) # tells the current location of pointer
    all_contents = f.read()
    f.write("*****")
    print(f.tell()) # tells the current location of pointer
    print(all_contents)
```

```python
with open("temp/temp_1_copy_2.py", "a+") as f:
    print(f.tell()) # tells the current location of pointer
    all_contents = f.read()
    f.write("print('hello world')")
    print(f.tell()) # tells the current location of pointer
    print(all_contents)
```

```python
with open("temp/temp_2.txt", "w+") as f:
    first_index = f.tell()
    print(first_index)

    f.write("hello world")
    
    last_index = f.tell()
    print(last_index)

    middle_index = (last_index // 2) + 1
    print(middle_index)

    f.seek(middle_index)
    f.write("****")
    print(f.tell())
    # print(all_contents)
```
