# control flow

```python
names = ["Sudharshan", "Praveen", "Abhijeet", "monal","Sandeep"]
```

```python
for name in names:
    if name == 'monal':
        continue
    name = name.capitalize()
    name = name.upper()
    print(name)
```

```python
for name in names:
    if name == 'monal':
        break
    name = name.capitalize()
    name = name.upper()
    print(name)
```

```python
count = 0
print("Start")
while True: # infinite
    if count>100 :
        break # control to exit based on condition
    print(count)
    count+=1
print("END")
```

```python
# Assignment
for i in range(1,11):
    print(i)
    for name in names:
        if name in ['Praveen', 'Abhijeet', 'monal', 'Sandeep']:
            break
        print(name)
```

```python
# Assignment

def add(a,b):
    print(a+b)

def substract(a,b):
    print(a-b)

def divide(a,b):
    print(a//b)

def multiply(a,b):
    print(a*b)

def calculator():
    choice = int(input("Enter 1 for add, 2 for subtract, 3 for divide, 4 for multiple : "))

    value_1 = float(input("Enter first value : "))
    value_2 = float(input("Enter Second value : "))

    if choice == 1:
        add(value_1, value_2)
    elif choice == 2:
        substract(value_1, value_2)
    elif choice == 3:
        divide(value_1, value_2)
    elif choice == 4:
        multiply(value_1, value_2)
    else:
        print("Please choose the correct option")
```

```python
calculator()
```
