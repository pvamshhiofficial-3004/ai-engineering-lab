# String operations

```python
name = "Monal S   "
```

```python
# strip()
name = name.strip() # it removes spaces from right and left side of the string
name
```

```python
# rstrip()
name = "   Monal S   "
# print(name.rstrip())
name.rstrip()
```

```python
# lstrip()
name = "   Monal S   "
name.lstrip()
```

```python
# remove space from both side without using strip()
name = "   Monal S   "
name = name.lstrip()
name = name.rstrip()
name
```

```python
name = "   Monal S   "
name = name.lstrip().rstrip()
name
```

```python
name = "   Monal S   "
name = name.rstrip().lstrip()
name
```

#### if-elif-else

```python
operation_name = "divide"

# contorl flow start
if operation_name == "add":
    print("Perform addition")
elif operation_name == "substract":
    print("Perform substraction")
elif operation_name == "divide":
    print("Perform division")
else:
    print("No selected operation name found.")
# contorl flow end
```

```python
operation_name = "add"

# contorl flow start
if operation_name == "add":
    print("Perform addition")
# contorl flow end

# contorl flow start
if operation_name == "add":
    print("Perform addition")
elif operation_name == "substract":
    print("Perform substraction")
elif operation_name == "divide":
    print("Perform division")
else:
    print("No selected operation name found.")
# contorl flow end

```
