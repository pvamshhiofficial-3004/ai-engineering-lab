# Example

```python
import time
```

```python
def my_decorator(func):

    def wrapper():
        t1 = time.perf_counter()
        func()
        t2 = time.perf_counter()
        total_time_takes = t2-t1
        print(f"Function took {total_time_takes:.2f} seconds")
    return wrapper
```

```python
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)
```

```python
loop_and_multiply = my_decorator(loop_and_multiply) # decorator with manual way
```

```python
loop_and_multiply()
```

```python
# decorator with syntactic sugar

# instead of writing

loop_and_multiply = my_decorator(loop_and_multiply) 

# Python gives shorthand way
@my_decorator
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)
```

```python
loop_and_multiply()
```

```python
# Step-1 : Create a decorator

def my_decorator(func):

    def wrapper():
        t1 = time.perf_counter()
        func()
        t2 = time.perf_counter()
        total_time_takes = t2-t1
        print(f"Function took {total_time_takes:.2f} seconds")
    return wrapper

# Step-2 : Create base function
@my_decorator
def loop_and_multiply():
    total = 1
    for i in range(1,100):
        total += (total*i)
        time.sleep(0.1)
    print(total)

# Step-3 : Using short hand way, add @my_decorator (@decorator_name) on top of your base function

```

```python
@my_decorator
def add():
    time.sleep(3)
    print("add")
```

```python
add()
```

# example

```python
def decorator_exep(func):
    def wrapper():
        try:
            func()
        except Exception as e:
            print(e)
    return wrapper
```

```python
@decorator_exep
def add():
    a = input("Enter a number : ")
    print(a + 100)
```

```python
add()
```

# Decorator Final concept

```python
def decorator_exep_v2(func):
    def wrapper(*args, **kwargs):
        try:
            func(*args,**kwargs)
        except Exception as e:
            print(e)
    return wrapper
```

```python
def add(*args, **kwargs):
    sum = 0
    for i in args:
        sum+=i
    for key, value in kwargs.items():
        sum+=value
    print(sum)
```

```python
add = decorator_exep_v2(add)
```

```python
add(1,b=2)
```

```python
add(1,b=2, c=100)
```

```python
add(1,2,3,4,5,6,b=2, c=100)
```

```python
# method - 2
@decorator_exep_v2
def add(*args, **kwargs):
    sum = 0
    for i in args:
        sum+=i
    for key, value in kwargs.items():
        sum+=value
    print(sum)
```

```python
add(1,2,3,4,5,6,b=2, c=100)
```

```python
# return something from wrapper

def decorator_exep_v3(func):
    def wrapper(*args, **kwargs):
        try:
            result = func(*args,**kwargs)
        except Exception as e:
            print(e)
        return result
    return wrapper
```

```python
@decorator_exep_v3
def add(*args, **kwargs):
    sum = 0
    for i in args:
        sum+=i
    for key, value in kwargs.items():
        sum+=value
    # print(sum)
    return sum
```

```python
result = add(1,2,3,4,5,6,b=2, c=100)
print(result)
```

# static methods : Used for logical group
- no object required
- cleaner name calling

```python
class MathUtils:

    @staticmethod
    def add(a,b):
        return a + b
    
    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a
```

```python
MathUtils.add(1,2)
```

```python
MathUtils.subtract_b_from_a(100, 2)
```

```python
class MathUtils_New:
     
    name = "UDSB2"

    @staticmethod
    def add(a,b):
        return a + b
    
    @staticmethod
    def subtract_a_from_b(a,b):
        return a - b

    @staticmethod
    def subtract_b_from_a(a,b):
        return b - a
    
    def class_method(self):
        print(self.add(1,2))
        print(MathUtils_New.add(1,2))
```

```python
MathUtils_New.add(1,2)
```

```python
obj_maths_1 = MathUtils_New()
```

```python
obj_maths_1.class_method()
```

```python
obj_maths_1.name = "GAI"
```

```python
obj_maths_1.name
```

```python
MathUtils_New.name
```
