```python
class OurBankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else :
            self.balance -= amount
        print(self.name,"withdrew :",amount,". New balance:", self.balance)
```

```python
user_1 = OurBankAccount("Monal", 10000)
```

```python
user_1.withdraw(10000)
```

```python
print(type(user_1))
```

```python
user_2 = OurBankAccount("deb", 10000)
```

```python
user_2.withdraw(10000)
```

# creating our own class

```python
# blue print
class Robot:
    # init or constructor is a special function that runs when you create object from class(blueprint)
    # init is optional
    # used to initialise or declare variables
    def __init__(a):
        print("Hello, welcome to robot!.")
        a.value_1 = 100
    
    def walk(a):
        print("Walking 10 meters.")
        print(a.value_1)
    
    def sing(a):
        print("Singing a song.")
        print(a.value_1)

```

```python
a = Robot() # returns object
```

```python
id(a)
```

```python
a.walk()
```

```python
a.sing()
```

```python
# blue print
class Robot:
    def __init__(self, customer_name=None):
        print("Hello, welcome to robot!.")
        self.name = "chitti"
        self.ram_in_gb = 10
        self.version = "0.0.1"
        print(customer_name)
        # self.customer_name = customer_name
    
    def config(self):
        print("Ram :", self.ram_in_gb)
        print("version :", self.version)
        print("Robot name :", self.name)
    
    def customer_details(self):
        print("Customer Name :", self.customer_name)

```

```python
r1 = Robot("monal")
```

```python
r1.config()
```

```python
r1.customer_details()
```

```python
r2 = Robot("Nikhil")
```

```python
r2.config()
```

```python
r2.customer_details()
```
