# Naming convention

- variable names can only contain letters, digits and underscore (_)
    - bmi, person0, person_1, _person, id_1
- it should not start with a digit 
- variable names are case sensitive : hello, Hello, heLlo
- avoid using python keywords : type, print, id, if, else, for

# Flow of python

in case of paranthesis, from inner brackets to outer
