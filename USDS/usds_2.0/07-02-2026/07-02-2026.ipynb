- Protected : {"syntax" : "_name", "Access" : "class + child classes"}
- Private : {"syntax" : "__name", "Access" : "only inside class"}

public, protected and private assumes the developer is responsible. Consent.

```python
# Scope levels in python

class BankAccount:
    def __init__ (self, name, balance):
        self.name = name # public
        self._account_type = "Savings" # Protected
        self.__balance = balance # private

    def get_balance(self):
        # getter()
        # to get value of the balance variable
        print("Your current balance :", self.__balance)
    
    def add_and_set_new_balance(self, amount):
        # setter()
        # to deposit money in our account
        self.__balance += amount
        print("Your current balance :", self.__balance)
```

# Inheritance

Inheritance allows a class to reuse properties and methods of another class

```python
# Single inheritance

class Device:
    def power_on(self):
        print("Device On")

class TV(Device):
    def show_channels(self):
        print("Channels 1., 2., 3.")
```

```python
tv_1 = TV()
```

```python
tv_1.show_channels()
```

```python
# MultiLevel

class Device:
    def a(self):
        print("abc")

class TV(Device):
    def a2(self):
        print("abc")

class SmartBulb(TV):
    def a3(self):
        print("abc")
```

# Example(hard)

```python
class MotherBoard:
    def __init__(self, ram, cpu_speed, cpu_type):
        self.ram = ram
        self.cpu_speed = cpu_speed
        self.cpu_type = cpu_type
    def start(self):
        print("Power flow started")
    def stop(self):
        print("Power flow stopped")
```

```python
class Computer(MotherBoard):
    def __init__(self, ram, cpu_speed, cpu_type):
        super().__init__(ram, cpu_speed, cpu_type)
        # MotherBoard(ram, cpu_speed, cpu_type)
        self.os = self.__check_os_and_set_os_type()
        # self.a = self.add(1,2)
        # a = add(1,2)
        # self.os = __check_os_and_set_os_type()

    def __check_os_and_set_os_type(self): # private method
        if self.cpu_type.lower() == "mac-chip":
            return "mac"
        else:
            return "other"
        
    def check_my_os(self): # public method
        print("User has :",self.os, "OS")
```

```python
cmp1 =Computer("12gb", "4.2ghz", "mac-chip")
```

```python
cmp1.check_my_os()
```

```python
cmp1.start()
```

```python
cmp1.stop()
```

```python
cmp1.ram
```

# Abstraction (Contract)

```python
from abc import ABC, abstractmethod
```

```python
class Payment(ABC): # requirement no 1

    @abstractmethod # decorator # requirement no 2
    def pay(self, amount, currency): # Arguments matching not-madatory
        pass
    
    @abstractmethod # requirement no 2
    def refund(self):
        pass

# contract called Payment has 2 method in the contract which all bank needs to implement
# first is pay
# second is refund
```

```python
class IPLDBank(Payment):

    def pay(self):
        print("PAY")

    def refund(self):
        print("REFUND")

    def new_user(self):
        print("hello")
```

```python
obj_b1 = IPLDBank()
```

```python
obj_b1.refund()
```

```python
obj_b1.pay()
```
